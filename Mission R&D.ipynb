```python
#Statistique sur nos donnes 
stats = X.describe().T

pourcentage = []
coef_variation = []

for col in X.columns:
    cellules_vides = np.mean(X[col].isnull())
    pourcentage.append(round(cellules_vides*100))
    moyenne = np.mean(X[col])
    ecart_type = np.std(X[col])
    coef_variation.append(round((ecart_type/moyenne)*100))
    #print('{} - {} - {} - {}% - {} - {} - {}%'.format(col, np.min(X[col]), np.max(X[col]), round(cellules_vides*100), moyenne, ecart_type, round((ecart_type/moyenne)*100)))

stats['donnees_manquantes %'] = pourcentage
stats['coef_variation %'] = coef_variation
stats
```

```python
features = ['Energie, Rglement UE N 1169/2011 (kcal/100 g)','Eau (g/100 g)','Protines, N x facteur de Jones (g/100 g)','Glucides (g/100 g)','Lipides (g/100 g)','Sucres (g/100 g)','Fructose (g/100 g)','Glucose (g/100 g)','Lactose (g/100 g)','Maltose (g/100 g)','Saccharose (g/100 g)','Amidon (g/100 g)','Fibres alimentaires (g/100 g)','Polyols totaux (g/100 g)','Cendres (g/100 g)','Alcool (g/100 g)','AG saturs (g/100 g)','AG monoinsaturs (g/100 g)','AG polyinsaturs (g/100 g)','Cholestrol (mg/100 g)','Sodium (mg/100 g)','Calcium (mg/100 g)','Fer (mg/100 g)','Magnsium (mg/100 g)','Phosphore (mg/100 g)','Potassium (mg/100 g)','Zinc (mg/100 g)','Vitamine D (g/100 g)','Vitamine E (mg/100 g)','Vitamine C (mg/100 g)','Vitamine B1 ou Thiamine (mg/100 g)','Vitamine B2 ou Riboflavine (mg/100 g)','Vitamine B12 (g/100 g)']

X = X[features]
X.shape
```

```python
X.columns
```

```python
for col in X.columns :
    X[col] = X[col].fillna(X[col].mean())
    
X
```

```python
# Min-Max Scaling (Normalisation)
X_normalized = (X - X.min()) / (X.max() - X.min())
X_normalized
```

```python
#Normalisation du dataset avec le Z score (standardization)
X_standardised = (X - X.mean(axis=0)) / X.std(axis = 0)
X_standardised
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2)
```

```python
#Linear SVC classification
```

```python
#Le modle du plus proche voisin
from sklearn.neighbors import KNeighborsClassifier

#Classification en groupes seulement reste en sous groupe et sous sous groupe

model = KNeighborsClassifier(n_neighbors=5) #varier les voisins pour avoir un score parfait

model.fit(X_train, y_train)
print('Score du training :', model.score(X_train, y_train))
print('Score du test :', model.score(X_test, y_test))

def nouvel_aliment(model, composants):
    x = np.array(composants).reshape(1, composants.shape[0])
    print(model.predict(x))
    print(model.predict_proba(x))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, model.predict(X_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=0)

from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve

def evaluation(model) :
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)

    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))

evaluation(model)
```

```python
#Cross validation : KFold
from sklearn.model_selection import  cross_val_score
from sklearn.model_selection import KFold, LeaveOneOut
cv = KFold(5,shuffle=True, random_state=0)
#cv = LeaveOneOut()
print(cross_val_score(KNeighborsClassifier(), X_normalized, y, cv=cv))
```
