```python
for colonne in df.columns:
    for ligne in range(df.shape[0]):
        if df[colonne][ligne] == '-' or df[colonne][ligne] == 'traces':
            df.loc[ligne, colonne] = np.nan
        if not pd.isnull(df[colonne][ligne]):
            if str(df.loc[ligne, colonne]).startswith('<'):
                df.loc[ligne, colonne] = str(df[colonne][ligne]).replace('<', '').replace('< ', '').replace(',', '.')
                df.loc[ligne, colonne] = np.nan 
            else:
                df.loc[ligne, colonne] = str(df[colonne][ligne]).replace(',', '.')
                df.loc[ligne, colonne] = float(df.loc[ligne, colonne])
```

```python
for colonne in features:
    df[colonne] = pd.to_numeric(df[colonne], downcast="float")
```
