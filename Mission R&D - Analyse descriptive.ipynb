```python
#Un peu de statistique sur nos données 
#On utilise ici le jeu de données qu'on a modifié
data = pd.read_excel('C:\\Users\\nisri\\Downloads\\final_dataset.xlsx')
stats = data.describe().T

pourcentage = []
coef_variation = []

for col in data.columns:
    cellules_vides = np.mean(data[col].isnull())
    pourcentage.append(round(cellules_vides*100))
    moyenne = np.mean(data[col])
    ecart_type = np.std(data[col])
    coef_variation.append(round((ecart_type/moyenne)*100))

stats['donnees_manquantes %'] = pourcentage
stats['coef_variation %'] = coef_variation


stats
```
