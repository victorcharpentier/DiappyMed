```python
import pandas as pd
import numpy as np
```

```python
#On charge nos données depuis le fichier excel
df = pd.read_excel('C:\\Users\\nisri\\Downloads\\ciqual_table.xls')
df.head()
```

```python
#Les features qu'on veut garder seront stockés en X
X = df.drop(['alim_grp_code', 'alim_ssgrp_code', 'alim_ssssgrp_code',
       'alim_grp_nom_fr', 'alim_ssgrp_nom_fr', 'alim_ssssgrp_nom_fr',
       'alim_code', 'alim_nom_fr', 'alim_nom_sci'], axis=1)
X
```

```python
#Les targets qu'on essaye d'obtenir
y = df[['alim_grp_code', 'alim_ssgrp_code', 'alim_ssssgrp_code']]
y
```

```python
#Replacer '-' par Nan
#Replacer 'traces' par Nan
#Replacer les '<?' par ?/2

for colonne in X.columns :
    for ligne in range(X.shape[0]):
        if X[colonne][ligne] == '-' or X[colonne][ligne] == 'traces':
            X.loc[ligne,colonne] = np.nan
        if not pd.isnull(X[colonne][ligne]) :
            if X.loc[ligne,colonne].startswith('<') :
                X.loc[ligne,colonne] = X[colonne][ligne].replace('<', '').replace('< ', '').replace(',', '.')
                X.loc[ligne,colonne] = float(X.loc[ligne,colonne])/2
            else :
                X.loc[ligne, colonne] = X[colonne][ligne].replace(',', '.')
                X.loc[ligne, colonne] = float(X.loc[ligne, colonne])
        #print(X.loc[ligne,colonne], 'de type ', type(X.loc[ligne,colonne]))
        #remarque : les nan sont des floats donc tout est de type str sauf les nans
        #Raison pour laquelle on est obligé de tout convertir en float

X
```

```python
#Imputation des données manquantes par une moyenne
for col in X.columns :
    X[col] = X[col].fillna(X[col].mean())
    
X
```

```python
# Min-Max Scaling (Normalisation)
X_normalized = (X - X.min()) / (X.max() - X.min())
X_normalized
```

```python
#Normalisation du dataset avec le Z score (standardization)
#X_standardised = (X - X.mean(axis=0)) / X.std(axis = 0)
#X_standardised
```

```python
#On concatène notre X et y pour avoir le dataset final
data = pd.concat([X_normalized, y], axis=1)
data
```
